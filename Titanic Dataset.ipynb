```python
from sklearn.linear_model import LogisticRegression
import pandas as pd
import numpy as np
from sklearn import preprocessing
data=pd.read_csv("training_titanic_x_y_train.csv")
df1=data.copy()
df1.drop(['Name'],axis=1,inplace=True)
df1.drop(['Ticket'],axis=1,inplace=True)
df1.drop(["Cabin"],axis=1,inplace=True)
df1.drop(["Survived"],axis=1,inplace=True)

df1.Embarked.dropna(inplace=True)
#df1.Age.fillna("0",inplace=True)
df1["Age"]=df1.groupby(['Pclass','Sex'])['Age'].apply(lambda x:x.fillna(x.median()))

df1.loc[df1['Sex']=='female','Sex']=0
df1.loc[df1['Sex']=='male','Sex']=1
df1.loc[df1['Embarked'].isna(),'Embarked']='S'
df1.loc[df1['Embarked']=='S','Embarked']=1
df1.loc[df1['Embarked']=='Q','Embarked']=2
df1.loc[df1['Embarked']=='C','Embarked']=3
df1
```

```python
df1.isnull().sum()
```

```python
data
```

```python
df=np.array(df1)
scaling=preprocessing.StandardScaler()
scaling.fit(df)
df=scaling.transform(df)
target=np.array(data['Survived'])
target
# from sklearn import model_selection
# X_train,X_test,Y_train,Y_test=model_selection.train_test_split(df,target)
```

```python
clf=LogisticRegression(multi_class='ovr')
clf.fit(df,target)
```

```python
clf.predict(df)
```

```python
clf.predict_proba(df)
```

```python
clf.score(df,target)
```
